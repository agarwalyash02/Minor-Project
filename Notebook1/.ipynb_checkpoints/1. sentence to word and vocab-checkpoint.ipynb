```python
array = df.values
```

```python
descriptions = {}

for x in array:
    image_name = x[0]
    
    if descriptions.get(image_name) is None:
        descriptions[image_name] = []
    descriptions[image_name].append(x[1])
```

```python
def removing_waste(sentence):
    sentence = str(sentence) #i have to add this as at some point error is coming not able to lower float value so converting all to string
    sentence = sentence.lower()
    sentence = re.sub("[^a-z]+"," ",sentence)
    sentence = sentence.split()
    sentence = [s for s in sentence if len(s)>1]
    sentence = " ".join(sentence)
    return sentence
```

```python
for key,cap in descriptions.items():
    for i in range(len(cap)):
        cap[i] = removing_waste(cap[i])
```

```python
total_words = []

for key in descriptions.keys():
    [total_words.append(i) for des in descriptions[key] for i in des.split()]
print("total Words %d"%len(total_words))
```

#### at first we have not added this step and our array size is very big

```python
#finding frequency of each word
freq = {}
for i in total_words:
    if freq.get(i) is None:
        freq[i] = 0
    freq[i]+=1
```

```python
len(freq)
```

```python
freq
```

```python
sorted_frq_cnt = sorted(freq.items(),reverse=True,key = lambda x:x[1]) #we have to use lambda as if wont use it will sort accoding with words

threshold = 10  #it is not fixed i have choosen to take 10 as if in 16,79,076 that word is coming only 10 times then it doesn't make any difference
sorted_frq_cnt = [x for x in sorted_frq_cnt if x[1] > threshold]
total_words = [x[0] for x in sorted_frq_cnt]
```

```python
print(len(total_words))
```

```python
total_words
```

```python
for key,cap in descriptions.items():
    for i in range(len(cap)):
        cap[i] = "startseq "+cap[i]+" endseq"
```

```python
descriptions['1000092795.jpg']
```

```python
word_to_idx = {}
idx_to_word = {}

for i,word in enumerate(total_words):
    word_to_idx[word] = i+1
    idx_to_word[i+1] = word
```

```python
print(len(idx_to_word))
print(len(word_to_idx))
```

```python
idx_to_word[1846] = 'startseq'
word_to_idx['startseq'] = 1846

idx_to_word[1847] = 'endseq'
word_to_idx['endseq'] = 1847

vocab_size = len(word_to_idx) + 1
print("vocab size",vocab_size)
```

```python
with open("../saved_models and files/word_to_idx.pkl","wb") as w2i:
    pickle.dump(word_to_idx,w2i)
with open("../saved_models and files/idx_to_word.pkl","wb") as i2w:
    pickle.dump(idx_to_word,i2w)
with open("../saved_models and files/description.pkl","wb") as tdes:
    pickle.dump(descriptions,tdes)
```
